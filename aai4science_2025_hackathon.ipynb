```python
import pandas as pd

# 1) Load the Excel
path = r"/home/hackthaon-data/AAI4ScienceHackathon2.xlsx"
xls = pd.ExcelFile(path)

print("Sheets:", xls.sheet_names)

# 2) Read the first sheet (change sheet_name if needed)
df = pd.read_excel(path, sheet_name=xls.sheet_names[0])

print("Shape:", df.shape)
print("Columns:")
for c in df.columns:
    print("-", c)

# 3) Pick the key columns (adjust names if they differ slightly)
cols_wanted = [
    "How many tasks did you complete?",
    "Save your personal github gist, different link than one provided and paste that link with responses here for evaluation",
    "Do you want to be considered for co-authorship if selected?",
    "Consent for research use",
    "Jay",
    "Choose 1 winner for your assignment"
]

missing = [c for c in cols_wanted if c not in df.columns]
if missing:
    print("\n⚠️ These columns were not found exactly as written:")
    for m in missing:
        print(" -", m)
    print("\nTip: copy/paste the exact column names from the printed list above.")
else:
    sub = df[cols_wanted].copy()
    print("\nPreview:")
    display(sub.head(10))

    # 4) Clean up a bit
    sub["How many tasks did you complete?"] = pd.to_numeric(
        sub["How many tasks did you complete?"], errors="coerce"
    )

    # 5) Simple stats
    print("\nTask completion stats:")
    print(sub["How many tasks did you complete?"].describe())

    print("\nCo-authorship interest counts:")
    print(sub["Do you want to be considered for co-authorship if selected?"].value_counts(dropna=False))

    print("\nConsent counts:")
    print(sub["Consent for research use"].value_counts(dropna=False))

    # 6) How many provided a gist link?
    gist_col = "Save your personal github gist, different link than one provided and paste that link with responses here for evaluation"
    has_gist = sub[gist_col].astype(str).str.contains("http", case=False, na=False)
    print("\nGist link provided:", has_gist.sum(), "out of", len(sub))
```
